```python
import requests,json

def current_weather():
    api_key = "bd59624c7ea5ec159aa3301437070458"
    requests_url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}"
    response = requests.get(requests_url)
    current_weather_data = response.json()
#     print(json.dumps(current_weather_data,indent=5))
    
    description = current_weather_data['weather'][0]['description']
    tempetature = current_weather_data['main']['temp']
    pressure = current_weather_data['main']['pressure']
    humidity = current_weather_data['main']['humidity']
    wind_speed = current_weather_data['wind']['speed']
    visibility = current_weather_data['visibility']
    city = current_weather_data['name']

    print("City: ",city)
    print("Description: ",description)
    print("Tempetature: ",tempetature)
    print("Pressure: ",pressure)
    print("Humidity: ",humidity)
    print("Wind Speed: ",wind_speed)
    print("Visibility: ",visibility)
        
current_weather()
```
